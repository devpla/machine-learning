### 2.4.1 factor형으로 변환

```r
m <- as.data.frame(lapply(mushroom, as.factor))
str(m)
```

```r
for (i in 1:length(mushroom)){
  mushroom[,i] <- factor(mushroom[,i])
}
```

### 2.4.2 훈련 및 테스트 데이터 만들기

```r
train_cnt <- 0.75 * dim(mushroom)[1]
idx <- sample(1:dim(mushroom)[1], train_cnt, replace = F)
mushroom_train <- mushroom[idx,]
mushroom_test <- mushroom[-idx,]
```

```r
NROW(mushroom_train)
```

```r
NROW(mushroom_test)
```

```r
table(mushroom_train$type)
```

```r
table(mushroom_test$type)
```

```r
model <- naiveBayes(mushroom_train[,-1], mushroom_train[,1])
result <- predict(model, mushroom_test[,-1])
```

```r
library(gmodels)
CrossTable(mushroom_test[,1], result)
```

### 2.4.3 라플라스 값 바꿔보기

```r
model <- naiveBayes(mushroom_train[,-1], mushroom_train[,1], laplace = 1)
result <- predict(model, mushroom_test[,-1])
CrossTable(mushroom_test[,1], result)
```

```r
model <- naiveBayes(mushroom_train[,-1], mushroom_train[,1], laplace = 0.1)
result <- predict(model, mushroom_test[,-1])
CrossTable(mushroom_test[,1], result)
```
