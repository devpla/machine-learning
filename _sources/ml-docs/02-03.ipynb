## 2.7 nltk.tokenize.word_tokenize

```python
survey = pd.read_csv("c:/data/naive_survey.csv", header = None, names = ['sentence', 'class'])
survey
```

```python
survey.info()
```

```python
allword = set([word for i in survey['sentence'] for word in word_tokenize(i)])
allword
```

```python
data = survey[['sentence','class']].values.tolist()
data
```

```python
t = [({word : (word in word_tokenize(x[0])) for word in allword}, x[1]) for x in data]
t
```

## 2.8 Naive Bayes 실행

```python
classifier = nltk.NaiveBayesClassifier.train(t)
classifier
```

```python
classifier.show_most_informative_features()
```

## 2.9 konlpy를 이용한 형태소 분석

```python
!pip install konlpy
```

```python
from konlpy.tag import Okt
```

```python
pos_tagger = Okt()
```

```python
test = '난 수업을 마치면 홍길동이랑 놀거야'
```

```python
pos_tagger.pos(test)
```

`[('난/Noun'), ('수업/Noun'), ('을/Josa'), ...]` 형식으로 바꾸기

```python
lst = []
for x, y in pos_tagger.pos(test):
  lst.append(x + '/' + y)
lst
```

```python
['/'.join(i) for i in pos_tagger.pos(test)]
```

```python
def tokenize(doc):
  return ['/'.join(i) for i in pos_tagger.pos(doc)]
```

```python
tokenize(test)
```

```python
data
```

```python
train_doc = [(tokenize(i[0]),i[1]) for i in data]
train_doc
```

```python
tokens = [j for i in train_doc for j in i[0]]
tokens
```

```python
def term_exists(doc):
  return {word : (word in set(doc)) for word in tokens}
```

```python
x_train = [(term_exists(d), c) for d, c in train_doc]
x_train
```

## 2.10 Naive Bayes 실행

```python
classifier = nltk.NaiveBayesClassifier.train(x_train)
classifier.show_most_informative_features()
```

```python
test = [('난 수업을 마치면 홍길동이랑 놀거야')]
test_doc = tokenize(test[0])
test_doc
```

```python
test_f = {word:(word in tokens) for word in test_doc}
classifier.classify(test_f)
```
